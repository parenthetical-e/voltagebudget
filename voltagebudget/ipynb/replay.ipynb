```python
from bokeh.plotting import figure
from bokeh.io import show, output_notebook
from bokeh.layouts import column, row
from bokeh.models import Range1d
output_notebook()

import numpy as np

from voltagebudget import util
from voltagebudget import neurons
from voltagebudget.exp.budget_experiments import replay
from fakespikes import util as fsutil

from pprint import pprint
```

```python
args = util.read_args("../data/fwd_test_ref_args.csv")
```

```python
ns, ts, budget = replay("../data/fwd_test_ref_args.csv", 
                        "../data/fwd_test_stim.csv", 
                        "../data/fwd_test.csv", 40, 30, verbose=True)

print(">>> {} spikes".format(ns.size))
```

```python
p = figure(plot_width=400, plot_height=200)
p.circle(ts, ns, color="black")
p.xaxis.axis_label = 'Time (s)'
p.yaxis.axis_label = 'N'
p.xgrid.grid_line_color = None
p.ygrid.grid_line_color = None
show(p)
```

```python
times = budget['times']
v = budget['V_m'][0, :]
m = times > 0.05  # Burn in time

p = figure(plot_width=400, plot_height=200)
p.line(x=times[m], y=v[m], color="black")
p.xaxis.axis_label = 'Time (s)'
p.yaxis.axis_label = 'Vm (volts)'
p.xgrid.grid_line_color = None
p.ygrid.grid_line_color = None
show(p)
```
