```python
from bokeh.plotting import figure
from bokeh.io import show, output_notebook
from bokeh.layouts import column, row
from bokeh.models import Range1d
from bokeh.io import export_png
output_notebook()

import numpy as np
```

```python
A = 1.
f = 31.

E = 2.323
d = -100e-3
t = np.linspace(0, 3, 2000)

phi0 = 0
x0 = A/2 * (1 + np.sin((t * f * 2 * np.pi) + phi0))

phi1 = (-(E + d) * 2 * np.pi * f) + np.pi/2
x1 = A/2 * (1 + np.sin((t * f * 2 * np.pi) + phi1))
```

```python
p = figure(plot_width=300, plot_height=200)
p.line(x=t, y=x0, color="grey", alpha=0.2)
p.line(x=t, y=x1, color="black", alpha=0.8)
p.line(x=E+d, y=np.linspace(0, 1, 100), color="red")
p.line(x=0, y=np.linspace(0, 1, 100), color="red", alpha=0.2)
p.line(x=t, y=np.repeat(A/2, t.size), color="grey")
show(p)
```
