```python
from bokeh.plotting import figure
from bokeh.io import show, output_notebook
from bokeh.layouts import column, row
from bokeh.models import Range1d
from bokeh.io import export_png
output_notebook()

import numpy as np

from voltagebudget.util import burst
from voltagebudget.util import find_phis
```

```python
t = 0.4
d = -2e-3
E = 0.22
f = 10
A = 0.1e-9
n_cycles = 1.5
phi_w, phi_E = find_phis(E, f, d, verbose=False)
time_step = 1e-5
```

```python
times, I_osc = burst((0, t), E , n_cycles, A, f, phi_w, time_step)
```

```python
p = figure(plot_width=400, plot_height=200)
p.circle(x=times, y=I_osc, color="black", alpha=0.1, size=0.2)
p.circle(x=times, y=A/2, color="grey", alpha=0.02, size=0.2)
p.xaxis.axis_label = 'Time (s)'
p.yaxis.axis_label = 'I_osc'
p.xgrid.grid_line_color = None
p.ygrid.grid_line_color = None
show(p)
```
