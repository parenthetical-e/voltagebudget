```python
from bokeh.plotting import figure
from bokeh.io import show, output_notebook
from bokeh.layouts import column, row
from bokeh.models import Range1d
output_notebook()

import numpy as np

from voltagebudget import util
from voltagebudget import neurons
from fakespikes import util as fsutil
```

# Run each model

Collecting a few random examples; save 'em.

```python
Nexamples = 10
f = 6
```

```python
mode = 'regular'
# adj = 0.1e-10
adj = 0.1e-10

# -
t = 0.5
# set mode
params, _, bias, sigma = util.read_modes(mode)

# -
ns_y, ts_y, budget_y = neurons.adex(Nexamples, t, np.asarray([0]), np.asarray([0]), 
                                  w_in=0, bias=bias*1.35, 
                                  A=.0, f=0,
                                  sigma=1.5e-9,
                                  step_params=(1.55e-10 + adj, 1, 1),
                                  **params)

ns_f, ts_f, budget_f = neurons.adex(Nexamples, t, np.asarray([0]), np.asarray([0]), 
                                  w_in=0, bias=bias, 
                                  A=.015e-9 + adj, f=f,
                                  sigma=1.5e-9,
                                  step_params=(1.475e-10 + adj, 1, 1),
                                  **params)

ns_s, ts_s, budget_s = neurons.adex(Nexamples, t, np.asarray([0]), np.asarray([0]), 
                                  w_in=0, bias=bias, 
                                  A=.045e-9 + adj, f=f,
                                  sigma=1.5e-9,
                                  step_params=(1.455e-10 + adj, 1, 1),
                                  **params)
```

```python
p = figure(plot_width=300, plot_height=120, tools='')
p.circle(ts_y, ns_y, color="black", size=0.4)
p.xaxis.axis_label = ''
p.yaxis.axis_label = 'N'
p.x_range = Range1d(0, t)
p.xgrid.grid_line_color = None
p.ygrid.grid_line_color = None
p.yaxis.major_tick_line_color = None  # turn off y-axis major ticks
p.yaxis.minor_tick_line_color = None  # turn off y-axis minor ticks
p.xaxis.major_tick_line_color = None  # turn off y-axis major ticks
p.xaxis.minor_tick_line_color = None  # turn off y-axis minor ticks
p.xaxis.major_label_text_font_size = '0pt'  # preferred method for removing tick labels
p.yaxis.major_label_text_font_size = '0pt'  # preferred method for removing tick labels
p.xaxis.visible = False
show(p)

p = figure(plot_width=300, plot_height=200, tools='')
for n in range(Nexamples):
    times = budget_y['times']
    v = budget_y['V_m'][n, :]

    p.line(x=times, y=v + n*19e-3, color="black", alpha=0.5)
    p.xaxis.axis_label = 'Time (s)'
    p.yaxis.axis_label = 'Vm (volts)'
    p.xgrid.grid_line_color = None
    p.ygrid.grid_line_color = None
    p.yaxis.major_tick_line_color = None  # turn off y-axis major ticks
    p.yaxis.minor_tick_line_color = None  # turn off y-axis minor ticks
#     p.xaxis.major_label_text_font_size = '0pt'  # preferred method for removing tick labels
    p.yaxis.major_label_text_font_size = '0pt'  # preferred method for removing tick labels

show(p)
```
