```python
from bokeh.plotting import figure
from bokeh.io import show, output_notebook
from bokeh.layouts import column, row
from bokeh.models import Range1d
output_notebook()

import numpy as np

from voltagebudget import util
from voltagebudget import neurons
from fakespikes import util as fsutil
```

# Set neurons mode

```python
t = 1.5

#set mode
mode = 'transient_spiking'
modes = util.read_modes()
params = modes[mode]

initial_inputs = params.pop('initial_inputs')
w_in = initial_inputs['w_in']
bias = initial_inputs['bias']
```

# Run a model

And observe the respones to a step current.

## Regular

```python
N = 1
ns_y, ts_y, budget = neurons.adex(N, t, np.asarray([0]), np.asarray([0]), 
                                  w_in=w_in, bias=bias, 
                                  sigma=0e-9, 
                                  A=.025e-9, f=0,
                                  step_params=(0.05e-10, 1, 0.5),
                                  **params)
```

```python
p = figure(plot_width=400, plot_height=200)
p.circle(ts_y, ns_y, color="black")
p.xaxis.axis_label = 'Time (s)'
p.yaxis.axis_label = 'N'
p.xgrid.grid_line_color = None
p.ygrid.grid_line_color = None
show(p)
```

```python
times = budget['times']
v = budget['V_m'][0, :]
# m = times > 0.05  # Burn in time

p = figure(plot_width=400, plot_height=200)
p.line(x=times, y=v, color="black")
p.xaxis.axis_label = 'Time (s)'
p.yaxis.axis_label = 'Vm (volts)'
p.xgrid.grid_line_color = None
p.ygrid.grid_line_color = None
show(p)
```
